Q1)Given below is a dictionary having two keys ‘Boys’ and ‘Girls’ and having two lists of heights of five Boys and
Five Girls respectively as values associated with these keys
Original dictionary of lists:
{'Boys': [72, 68, 70, 69, 74], 'Girls': [63, 65, 69, 62, 61]}
From the given dictionary of lists create the following list of dictionaries:
[{'Boys': 72, 'Girls': 63}, {'Boys': 68, 'Girls': 65}, {'Boys': 70, 'Girls': 69}, {'Boys': 69, 'Girls': 62}, {‘Boys’:74,
‘Girls’:61]

```python
dic1={'Boys': [72, 68, 70, 69, 74], 'Girls': [63, 65, 69, 62, 61]}
```

```python
listBoys=dic1['Boys']
listGirls=dic1['Girls']
```

```python
res=[]
for i in range(len(listBoys)):
    dic2={}
    dic2['Boys']=listBoys[i]
    dic2['Girls']=listGirls[i]
    res.append(dic2)
res
```

# Q5

```python
from sklearn import datasets
ir = datasets.load_iris()
iris = pd.DataFrame(data=ir.data, columns=ir.feature_names)
iris["Species"] = ir.target
iris.replace({"Species": {0: "Iris-setosa",1:'Iris-versicolor',2:'Iris-virginica'}},inplace=True)
iris
```

a. Plot bar chart to show the frequency of each class label in the data.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.bar(iris['Species'].value_counts().index,height=iris['Species'].value_counts(),color=['b','g','r'])
plt.xlabel("Species")
plt.ylabel('Species Counts')
plt.title('Species vs Frequency')
plt.show()
```

b. Draw a scatter plot for Petal width vs sepal width.

```python
plt.scatter(iris["petal width (cm)"],iris['sepal width (cm)'])
plt.xlabel("Petal Width")
plt.ylabel("Sepal Width")
plt.title("Petal Width vs Sepal Width")
plt.show()
```

c. Plot density distribution for feature petal length.

```python
iris['petal length (cm)'].plot.density()
plt.title('Density distribution of petal length')
plt.show()
```

d. Use a pair plot to show pairwise bivariate distribution in the Iris Dataset

```python
sns.pairplot(iris)
plt.show()
```
